```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
```

```python
scale = StandardScaler()
```

```python
df_scale = scale.fit_transform(df)
```

```python
#KMeans
#métricas para ver se é bom ou ruim: Elbow (técnica do cotovelo), Silhouette Score
#se não houver métricas, verificar com a área de negócios se faz sentido o resultado
#objetivo do agrupamento é criar um grupo que possua a menor distancia intracluster e a maior distancia intercluster (grupos mais uniformes)

centroides = [2,3,4,5,6,7,8,9,10]

for num_centroide in centroides:
    km = KMeans(n_clusters=num_centroide)
    predicao = km.fit_predict(df_scale)

    score = silhouette_score(df_scale, predicao)
    print(f'O valor de silhouette_score é {score}, para um numero de clusters igual a {num_centroide}')
    

#Melhor corte é 2, está dando o maior silhouette_score. Mas tem que ser verificado com a área de negócios
```

```python
#Agglometarive

clusters = [2,3,4,5,6,7,8,9,10]

for num_clusters in clusters:
    agg = AgglomerativeClustering(n_clusters=num_clusters)
    predicao = agg.fit_predict(df_scale)

    score = silhouette_score(df_scale, predicao)
    print(f'O valor de silhouette_score é {score}, para um numero de clusters igual a {num_clusters}')
```

```python
#DBSCAN - Density Based Clustering

eps = [5,7,9,11,15]

for num_eps in eps:
  db = DBSCAN(eps=num_eps)
  predicao = db.fit_predict(df_scale)
  score = silhouette_score(df_scale, predicao)
  print("O valor de silhouette_score é {}, para um numero de clusters igual a {}".format(score, num_eps))

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
inertia_list = []
for i in range(2,8):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(df_scale)
  inertia_list.append(kmeans.inertia_)
        
# plotting
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
sns.lineplot(y=inertia_list, x=range(2,8), ax=ax)
ax.set_xlabel("Cluster")
ax.set_ylabel("Inertia")
ax.set_xticks(list(range(2,8)))
fig.show()
```
